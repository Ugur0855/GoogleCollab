```
from skimage import data, io, filters, feature, exposure
from skimage.transform import rotate
from skimage.morphology import closing, square
from skimage.color import rgb2gray
from skimage.io as io
```

```
I = data.coins()
io.imshow(I)
```

```
kenarlar = filters.sobel(I);
io.imshow(kenarlar);
```

```
edges1 = feature.canny(I, sigma=4)
io.imshow(edges1);
```

```
IR = rotate(I, 30, True)
io.imshow(IR);
```

```
I = data.moon()
I_eq = exposure.equalize_hist(I)
io.imshow(I)
io.show()
io.imshow(I_eq)
```

```
I = 1.0 - rgb2gray(data.horse())
io.imshow(I)
io.show()
I2 = closing(I, square(50))
io.imshow(I2)
```

```
mandrill = io.imread("mandrill.png")
io.imshow(mandrill)
```
