```
from skimage import data, io, filters, feature, exposure
from skimage.transform import rotate
from skimage.morphology import closing, square
from skimage.color import rgb2gray

import numpy as np
```

# Resim görüntüleme ve kenar filtresi

```
I = data.coins()
io.imshow(I)
```

```
kenarlar = filters.sobel(I);
io.imshow(kenarlar);
```

# Canny filtresi ve parametre değiştirme

```
edges1 = feature.canny(I, sigma=4)
io.imshow(edges1);
```

# Görüntü döndürme

```
IR = rotate(I, 30, True)
io.imshow(IR);
```

# Histogram eşitleme

```
I = data.moon()
I_eq = exposure.equalize_hist(I)
io.imshow(I)
io.show()
io.imshow(I_eq)
```

# Morfolojik işlemler

```
I = 1.0 - rgb2gray(data.horse())
io.imshow(I)
io.show()
I2 = closing(I, square(50))
io.imshow(I2)
```

# Görüntü dosyası okuma piksellere erişim

```
mandrill = io.imread("/content/drive/MyDrive/Colab Notebooks/mandrill.png")
io.imshow(mandrill)
io.show()
n_ims = len(mandrill)
for i in range (1,225):
  for j in range (1, 225):
    mandrill[i,j] = mandrill[i,j] + 50

io.imshow(mandrill)
```
